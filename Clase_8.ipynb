```python
gold = pd.read_csv("goldprice.csv")
```

```python
gold.info()
```

```python
gold["Date"] = pd.to_datetime(gold["Date"])
```

```python
gold.info()
```

```python
# 1 - Crea columna Year

gold["Year"] = gold.Date.dt.year
```

```python
gold.info()
```

```python
gold
```

```python
# 2 - Determinar el precio promedio por año

gold.set_index("Date", inplace=True)
    
```

```python
gold.groupby(gold.index.year)["Price"].mean()
```

```python
# 3 - Graficar la información anterior en Gráfico de Barras

promedio_anio = gold.groupby(gold.index.year)["Price"].mean()
```

```python
promedio_anio.plot(kind="bar")
```

```python
# 4 - Calcular el año de mayor promedio y el de menor promedio

print(gold[["Year", "Price"]].groupby("Year").mean().idxmax())
print(gold[["Year", "Price"]].groupby("Year").mean().max())
```

```python
print(gold[["Year", "Price"]].groupby("Year").mean().idxmin())
print(gold[["Year", "Price"]].groupby("Year").mean().min())
```

```python
# Evolución del promedio por año (Graficar)

promedio_anio.plot(kind="line")
```

## Ejercicio de ventas de una tienda

```python
df = pd.read_csv("01_ventas_tienda.csv", parse_dates=["Fecha"])
```

```python
df
```

```python
df.info()
```

## 🧩 Consignas

1. Mostrar las **primeras 5 filas** del dataset.  
2. Calcular una nueva columna llamada `Total` = `PrecioUnitario * Cantidad`.  
3. Calcular las **ventas totales por cliente**.  
4. Determinar **qué producto generó mayor ingreso total**.  
5. Calcular la **cantidad total de productos vendidos por categoría**.  
6. Obtener el **promedio de gasto por venta**.  
7. Filtrar las ventas de **febrero** y calcular su ingreso total.  
8. Mostrar un resumen de **estadísticas descriptivas** de `PrecioUnitario`, `Cantidad` y `Total`.  
9. (Extra) Graficar:
   - Top 10 clientes por ingresos
   - Ingresos por producto (Top 10)
   - Cantidad vendida por categoría
   - Distribución del ticket (`Total`)
   - Ingresos por mes (serie temporal)

```python
# 1. 

df.head(5)
```

```python
# 2.

df["Total"] = df["PrecioUnitario"] * df["Cantidad"]
```

```python
df
```

```python
df
```

```python
ventas_x_cliente = df.groupby("Cliente")["Total"].sum().sort_values(ascending=False)
```

```python
ventas_x_cliente
```

```python
# 4.

ingresos_x_producto = df.groupby("Producto")["Total"].sum().sort_values(ascending=False)
```

```python
ingresos_x_producto
```

```python
# 5.

cantidad_x_categoria = df.groupby("Categoría")["Cantidad"].sum().sort_values(ascending=False)
```

```python
cantidad_x_categoria
```

```python
# 6.

promedio_de_gasto = df["Total"].mean()
```

```python
promedio_de_gasto
```

```python
# 7.
ventas_de_febrero = df[ df["Fecha"].dt.month == 2 ]["Total"].sum()
```

```python
ventas_de_febrero
```

```python
# 8.

df[["PrecioUnitario", "Cantidad", "Total"]].describe()
```

```python
"""
9. (Extra) Graficar:
   - Top 10 clientes por ingresos
   - Ingresos por producto (Top 10)
   - Cantidad vendida por categoría
   - Distribución del ticket (`Total`)
   - Ingresos por mes (serie temporal)
"""

ventas_x_cliente.head(10).plot(kind="bar")
```

```python
ingresos_x_producto.head(10).plot(kind="barh")
```
