## Funciones de Orden Superior

```python
def ecu1(x):
    return x**2

def ecu2(x):
    return x**3

# Funcion de Orden Superior: porque recibe el nombre de otra funcion
def aplicar(funcion, lista):
    resultado = []
    for i in lista:
        resultado.append(funcion(i))

    return resultado

numeros = [2,5,3,7]

r = aplicar(ecu2, numeros)
print(r)
```

## Funciones built-in

1. map
2. filter
3. sorted

```python
def cuadrado(x):
    return x**2

def cubo(x):
    return x**3

numeros = [2,5,3,7]

resultado = map(cuadrado, numeros)
print(list(resultado))
```

```python
anios = [1800, 1983, 1980, 2001, 1996, 1840, 2020, 2026]

def bisiesto(anio):
    if anio%400==0:
        return True
    elif anio%4==0 and anio%100!=0:
        return True
    return False

resultado = filter(bisiesto, anios)
print(tuple(resultado))
```

```python
# Ordenar por Apellido 

nombres = ["Juan Perez", "Ana Zaraza", "Pedro Gonzalez", "Maria Ricardi", "Alejandro Alonso"]

sorted(nombres, key=lambda x:x.split()[-1], reverse=True)
```
